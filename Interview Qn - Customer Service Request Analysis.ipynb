#### 4.2 Sort the complaint types based on the average Request_Closing_Time grouping them for different locations

```python
data['created_date']=pd.to_datetime(data['created_date'])
data['closed_date']=pd.to_datetime(data['closed_date'])
data['average_closing_time']=data['closed_date']-data['created_date']
```

```python
data.sample()
```

```python
data['average_closing_time']=(data['closed_date']-data['created_date']).dt.total_seconds() / 3600
```

```python
data.sample()
```

```python
complaint_avrg_rqst = data.groupby(['complaint_type','city'])['average_closing_time'].mean().reset_index()
sorted_list=complaint_avrg_rqst.sort_values(by='average_closing_time')
sorted_list
```
