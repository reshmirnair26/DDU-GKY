## Creating Data Frames

```python
import pandas as pd
```

### From Dictionary

```python
dict01 = {'Age':[10,20,15], 'Name':['Vinu', 'Anu', 'Manu']}
```

```python
df2=pd.DataFrame(dict01)
```

```python
df2
```

```python
df2.info()
```

### Data Frame using Lists

```python
names = ['Varun', 'Arun', 'Kiran']
age = [25,28,30]
```

```python
df1 = pd.DataFrame(list(zip(names, age)), columns=['Names', 'Age'])
```

```python
df1
```

### Data Frame from Lists of Dictionaries

```python
lst = [{'Name':'Arun', 'Age':29, 'Gender':'Male'},
       {'Name':'Manu', 'Age':28, 'Gender':'Male'},
       {'Name':'Safeer', 'Age':20, 'Gender':'Male'},
       {'Name':'Radha', 'Age':27, 'Gender':'Female'},
       {'Name':'Vivek', 'Age':28, 'Gender':'Male'}]
```
